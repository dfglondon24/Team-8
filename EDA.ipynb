```python
# List of sheet names as per user description
sheet_names = [
    'Basic Schs 2010_2011', 'Basic Schs 2011_2012', 'Basic Schs 2012_2013',
    'Basic 2013_2014', 'Basic_2014_2015', 'Basic_2015_2016',
    'SHS 2010_2011', 'SHS 2011_2012', 'SHS 2012_2013', 'SHS 2013_2014', 'SHS_2014_2015', 'SHS_2015_2016',
    'TVET 2010_2011', 'TVET 2011_2012', 'TVET 2012_2013', 'TVET 2013_2014', 'TVET_2014_2015', 'TVET_2015_2016'
]

# Separating sheets into Basic, SHS, and TVET categories
basic_sheets = [sheet for sheet in sheet_names if 'Basic' in sheet]
shs_sheets = [sheet for sheet in sheet_names if 'SHS' in sheet]
tvet_sheets = [sheet for sheet in sheet_names if 'TVET' in sheet]

# Load and concatenate the relevant sheets into separate dataframes
df_basic = pd.concat([pd.read_excel(xls, sheet_name=sheet) for sheet in basic_sheets], ignore_index=True)
df_shs = pd.concat([pd.read_excel(xls, sheet_name=sheet) for sheet in shs_sheets], ignore_index=True)
df_tvet = pd.concat([pd.read_excel(xls, sheet_name=sheet) for sheet in tvet_sheets], ignore_index=True)

# Display the shapes of the combined dataframes for each category
df_basic.shape, df_shs.shape, df_tvet.shape
```

```python
df_basic.head()
```

```python
# To analyze the gender breakdown of students across all classes (public/private) over time, we need to filter out the relevant columns.
# We'll look for columns with 'B' (Boys) and 'G' (Girls) in their names, which represent the gender breakdown.

# Extract columns that contain 'B' or 'G' (for Boys and Girls), and public/private ('Pub'/'Prv')
gender_columns = [col for col in df_basic.columns if ('B_' in col or 'G_' in col) and ('Pub' in col or 'Prv' in col)]

# Creating a new dataframe with only the gender-related columns
df_gender_breakdown = df_basic[['Region', 'District', 'Academic_Year'] + gender_columns]

# Grouping by year to analyze trends over time for public and private schools
gender_trend = df_gender_breakdown.groupby('Academic_Year').sum()

# # Displaying the trend data for gender breakdown across the years
# import ace_tools as tools; tools.display_dataframe_to_user(name="Gender Breakdown Trend", dataframe=gender_trend)

gender_trend.head()
```

```python
# Finding columns that still have any NaN values in the original dataframes (before preprocessing)
# We'll check the original combined datasets for columns that contain NaN values.

nan_columns_basic = df_basic.columns[df_basic.isnull().any()].tolist()
nan_columns_shs = df_shs.columns[df_shs.isnull().any()].tolist()
nan_columns_tvet = df_tvet.columns[df_tvet.isnull().any()].tolist()

# Displaying the columns that contain NaN values in each dataset
nan_columns_basic, nan_columns_shs, nan_columns_tvet
```

```python
# Dropping columns with NaN values from each dataset
df_basic_cleaned = df_basic.dropna(axis=1, how='any')
df_shs_cleaned = df_shs.dropna(axis=1, how='any')
df_tvet_cleaned = df_tvet.dropna(axis=1, how='any')

# Display the shapes of the cleaned dataframes to verify the changes
df_basic_cleaned.shape, df_shs_cleaned.shape, df_tvet_cleaned.shape
```

```python
# Selecting only the numeric columns from the combined datasets for preprocessing
# First, we will identify the numeric columns and then handle missing values and other potential preprocessing steps.

# For Basic Schools Data
df_basic_numeric = df_basic.select_dtypes(include=['float64', 'int64'])

# For SHS Data
df_shs_numeric = df_shs.select_dtypes(include=['float64', 'int64'])

# For TVET Data
df_tvet_numeric = df_tvet.select_dtypes(include=['float64', 'int64'])

# Handling missing values in numeric columns
# Filling missing values with the mean of the respective columns for each dataset

df_basic_numeric.fillna(0)
df_shs_numeric.fillna(0)
df_tvet_numeric.fillna(0)

# Display the shapes of the preprocessed numeric data
df_basic_numeric.shape, df_shs_numeric.shape, df_tvet_numeric.shape
```

```python
# Creating a simplified dataset with total counts for public and private schools by Academic Year and Region
df_simplified = df_combined_basic_cleaned.groupby(['Academic_Year', 'Region']).agg({
    'Cr_TotSchs_Pub': 'sum',
    'Kg_TotSchs_Pub': 'sum',
    'Prm_TotSchs_Pub': 'sum',
    'JHS_TotSchs_Pub': 'sum',
    'Cr_TotSchs_Prv': 'sum',
    'Kg_TotSchs_Prv': 'sum',
    'Prm_TotSchs_Prv': 'sum',
    'JHS_TotSchs_Prv': 'sum'
}).reset_index()

# Adding a 'Public' and 'Private' summary column
df_simplified['Total_Public_Schools'] = df_simplified[['Cr_TotSchs_Pub', 'Kg_TotSchs_Pub', 'Prm_TotSchs_Pub', 'JHS_TotSchs_Pub']].sum(axis=1)
df_simplified['Total_Private_Schools'] = df_simplified[['Cr_TotSchs_Prv', 'Kg_TotSchs_Prv', 'Prm_TotSchs_Prv', 'JHS_TotSchs_Prv']].sum(axis=1)

# Keeping only the relevant columns: Academic Year, Region, Public and Private Schools
df_simplified = df_simplified[['Academic_Year', 'Region', 'Total_Public_Schools', 'Total_Private_Schools']]

# Displaying the first few rows of the simplified dataset
df_simplified.head()
```

```python
# Plotting the year-wise trend for Public schools by region
plt.figure(figsize=(14, 6))
for region in df_simplified['Region'].unique():
    region_data = df_simplified[df_simplified['Region'] == region]
    plt.plot(region_data['Academic_Year'], region_data['Total_Public_Schools'], marker='o', label=f"{region} - Public")

plt.title('Year-wise Trend for Public Schools by Region', fontsize=16)
plt.xlabel('Academic Year', fontsize=12)
plt.ylabel('Total Public Schools', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Plotting the year-wise trend for Private schools by region
plt.figure(figsize=(14, 6))
for region in df_simplified['Region'].unique():
    region_data = df_simplified[df_simplified['Region'] == region]
    plt.plot(region_data['Academic_Year'], region_data['Total_Private_Schools'], marker='o', label=f"{region} - Private")

plt.title('Year-wise Trend for Private Schools by Region', fontsize=16)
plt.xlabel('Academic Year', fontsize=12)
plt.ylabel('Total Private Schools', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Displaying the plots
plt.show()

# plot for the total number of schools
```
