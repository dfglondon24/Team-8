```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('data/Data/AfriKids/InternalData/Project_Surveys/All AfriKids Projects Beneficiary Monitoring for 2018.xlsx')

# Initialize a directed graph
G = nx.DiGraph()

# Add nodes and edges
for index, row in df.iterrows():
    school_node = f"School: {row['Bongo school']}"  # Customize this
    G.add_node(school_node)
    
    student_node = f"Student: {index} - Age: {row['Age']} - Sex: {row['Sex']}"
    G.add_node(student_node)
    
    G.add_edge(student_node, school_node)

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True)
plt.title("Relational Graph of Students and Schools")
plt.show()
```
