```python
import pandas as pd

def read_excel_sheet(file_path, sheet_name=None):
    """
    Reads a specific sheet from an Excel file and returns a DataFrame.
    
    Args:
    file_path (str): Path to the Excel file.
    sheet_name (str or int, optional): Name or index of the sheet to read. If None, reads the first sheet.
    header_row (int, optional): Row number to use as the header (0-indexed). 
                                If None, assumes the first row is the header.
    
    Returns:
    pandas.DataFrame: DataFrame containing the sheet data.
    """
    # If header_row is None, pandas will automatically use the first row as header
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    return df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def describe_dataframe(df):
    """Provide a comprehensive description of the dataframe."""
    print(df.info())
    print("\nNumeric Columns Summary:")
    print(df.describe())
    print("\nNon-Numeric Columns Summary:")
    print(df.describe(include=['O']))
    print("\nMissing Values:")
    print(df.isnull().sum())

def plot_numeric_relationship(df, x, y, hue=None, plot_type='scatter', figsize=(10, 6)):
    """
    Plot relationship between two numeric columns using scatter or line plot.
    
    Args:
    df (pd.DataFrame): The dataframe containing the data
    x (str): Column name for x-axis
    y (str): Column name for y-axis
    hue (str, optional): Column name for color encoding
    plot_type (str): Type of plot - 'scatter' or 'line'
    figsize (tuple): Figure size
    
    Returns:
    None (displays the plot)
    """
    plt.figure(figsize=figsize)
    
    if plot_type == 'scatter':
        sns.scatterplot(x=x, y=y, hue=hue, data=df)
        plot_type_name = 'Scatter plot'
    elif plot_type == 'line':
        sns.lineplot(x=x, y=y, hue=hue, data=df)
        plot_type_name = 'Line plot'
    else:
        raise ValueError("plot_type must be either 'scatter' or 'line'")
    
    plt.title(f'{plot_type_name} of {y} vs {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    
    if hue:
        plt.legend(title=hue, bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
    
    plt.show()



# Sample usage
# describe_dataframe(data)
# plot_numeric_relationship(data, 'JHS_TotSchs', 'Prm_TotSchs', plot_type='line')
```
